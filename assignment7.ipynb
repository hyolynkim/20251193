14. B-트리에서 리프가 아닌 노드에 있는 원소의 직후 원소는 항상 리프 노드에 있다. 왜 그런가?
-> B-트리에서 특정 원소의 직후 원소는 해당 원소의 오른쪽 서브트리에서 가장 작은 값을 의미한다. 리프가 아닌 노드에서 오른쪽 자식 노드로 내려가면, 그 이후로는 가장 왼쪽 자식 포인터를 따라 최하단 리프 노드에 도달할 때까지 계속 내려가게된다. 탐색이 멈추는 지점은 항상 더 이상 자식이 없는 리프 노드이며, 그곳의 가장 첫 번째 원소가 바로 직후 원소가 된다.따라서 구조적으로 모든 내부 노드의 직후 원소는 항상 리프 노드에 존재하게 된다.
