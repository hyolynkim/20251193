1) 3개 (리프노드= 자식이 없는 노드)
2) 가능한 리프 노드를 최대로 만들기 위해서는 포화이진트리를 만들어야한다. 따라서 가능한 리프 노드의 최대 개수는 8
개이다.
3) 1개 , 포화이진트리가 되었을 때 높이가 4인 이진 검색트리의 리프 노드는 8개, 내부노드는 7개임으로 8-7=1 , 차이는 1개가 난다.
4) 그렇다. 포화 이진 트리에서 리프 노드는 2^h, 내부 노드는 2^h - 1의 개수를 가지므로 수식적으로 그 차이는 항상 1이 된다.
