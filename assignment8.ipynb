2-(1) **해시 테이블 크기를 1,204로 잡고 해시 함수로 h(x) = x%1024를 사용하였다. 어떤 결점이 있는가?**

-> 해시 함수에서 1,024($2^{10}$)처럼 2의 거듭제곱수를 나누는 값으로 사용하면, 결과값이 데이터의 하위 비트들에 의해서만 결정되는 문제가 발생한다. 이 경우 데이터의 뒷부분 모양이 비슷하면 해시 결과가 똑같이 나오기 쉬워, 특정 자리에 데이터가 몰리는 충돌 현상이 빈번해질 수 있다. 이를 해결하기 위해서는 나누는 값을 2의 거듭제곱수가 아닌 소수로 설정해야 데이터가 테이블 전체에 골고루 퍼지게 된다.

2-(2) **한 학생이 “다음과 같이 해시 함수를 만들면 확률적으로 고른 분포를 보이는 좋은 함수가 되지 않을까요?”라고 물었다. 단, random(0,10)은 0보다 크고 10 이하인 임의의 실수를 생성한다.
 h(x) = x*random(0,10)%m 이것은 해시 함수가 될 수 없다. 이유를 설명하시오.**

-> 해시 함수는 동일한 입력값에 대해 항상 동일한 결과값을 내놓아야 하지만, 이 함수는 실행할 때마다 결과가 계속 변한다. 만약 데이터 저장 시점과 검색 시점의 결과가 달라지면, 저장된 데이터를 다시 찾아낼 수 없게 되기 때문에 위의 함수는 해싱함수가 될 수 없다.

2-(7) **체이닝을 사용하는 충돌 해결 방법에서 각 연결 리스트에 새로운 키가 삽입될 때 리스트의 맨 앞에 삽입하지 않고 맨 뒤에 삽입한다면 효율성 면에서 어떤 변화가 생기는가?**

-> 연결 리스트의 맨 뒤에 데이터를 넣으려면 리스트 전체를 처음부터 끝까지 훑으며 마지막 노드를 찾아야 하므로 삽입 속도가 느려진다. 반면 맨 앞에 삽입하면 리스트의 길이에 상관없이 즉시 데이터를 추가할 수 있어 훨씬 효율적이다.

2-(8) **체이닝을 사용하는 충돌 해결 방법에서 각 연결 리스트를 정렬된 형태로 유지한다면 효율성 면에서 어떤 변화가 생기는가? (삽입, 삭제, 성공적인 검색, 실패하는 검색의경우로 나누어 작성하시오.)**

-> 삽입의 경우는 새로운 키를 넣을 때 적절한 위치를 찾아야 하므로, 리스트 끝까지 가지 않아도 됐던 이전과 달리 평균적인 탐색 시간이 늘어나 효율이 떨어질 수 있다.

삭제 및 성공적인 검색의 경우는 둘다 리스트가 정렬되어 있더라도 특정 데이터를 찾으려면 어차피 해당 위치까지 가야 하므로, 정렬되지 않았을 때와 비교해 시간 효율에 큰 차이가 없다.

실패하는 검색의 경우에는 찾으려는 값보다 큰 값이 나오는 순간 검색을 중단할 수 있어, 리스트 끝까지 모두 확인해야 했던 정렬 전보다 검색 속도가 빨라질 수 있다.
